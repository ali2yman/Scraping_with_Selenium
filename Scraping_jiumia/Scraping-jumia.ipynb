```python
import time
import requests
import os
import pandas as pd 
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome()
driver.get('https://www.jumia.com.eg/catalog/?q=jacket+for+mens')
```

```python
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
time.sleep(5)   # Allow 3 seconds for the web page to load 


product_data = []

products = driver.find_elements("xpath", '//div[@class="info"]')  # Parent div for each product

for product in products:
    product_data.append(product.text)
```

```python
products = []

for item in product_data:
    # e7na momken nsplit kol element using "\n" to get the data we need
    parts = item.split("\n")

    # hna ana 3mlt description = parts[0]  # 3shan da akeed hykon ,wgoud aya haga tanya momken mtkonsh mwgoda f n5ly el defult bta3ha None
    description = parts[0]
    price_after = None
    price_before = None
    discount = None
    rating = None

    for part in parts[1:]:

        if "EGP" in part:
            # Check if the part contains " - " so this is price after discount
            if " - " in part:
                price_after = part
            # if price after still None >> this means there is no -   >>> so get the price and put it in price after
            elif price_after is None:
                price_after = part
            # this happen when price there is EGP and price after has a value 
            else:
                price_before = part
        elif "%" in part:
            discount = part
        elif "out of 5" in part:
            rating = part.split(" out of 5")[0]


    products.append({
        "Description": description,
        "Price After Discount": price_after,
        "Price Before Discount": price_before,
        "Discount Percentage": discount,
        "Rating": rating
    })

# Convert to DataFrame
df = pd.DataFrame(products)

# Display DataFrame
df.head()
```

```python
image_links = driver.find_elements("xpath",'//div[@class="img-c"]/img[@class="img"]')

image_urls = []

for i in image_links:
    image_urls.append(i.get_attribute("src"))


len(image_urls)
```

```python
df["image_urls"]  = image_urls
df.head()
```

```python
df.to_csv("products.csv",index=False)   
```

```python
output_dir = "downloaded_images"
# Create directory if it doesn't exist >>  y3ni lw file mwgod 5las msh hy3ml create
os.makedirs(output_dir, exist_ok=True) 

for index, url in enumerate(image_urls, start=1):
    try:
        # Fetch the image
        # timeout >> waits 10 seconds to respond >> if there is no replay >> raise an error 
        response = requests.get(url, timeout=10)  #
        response.raise_for_status()  
        
        # Extract image extension (e.g., .jpg, .png)
        ext = os.path.splitext(url)[-1]  # Get the file extension from the URL
        if ext not in [".jpg", ".jpeg", ".png", ".gif", ".bmp", ".webp"]:
            ext = ".jpg"  #  make it with a defult extension 
        
        # Save the image
        file_name = f"image_{index}{ext}"  # Unique name for each image
        file_path = os.path.join(output_dir, file_name)
        
        with open(file_path, "wb") as file:
            file.write(response.content)
        
        print(f"Saved: {file_path}")
    
    except requests.RequestException as e:
        print(f"Failed to download {url}: {e}")
```
